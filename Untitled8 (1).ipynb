```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import folium
from folium.plugins import HeatMap
```

```python
df = pd.read_csv("signal_metrics.csv")
```

```python
df.head()
```

```python
#calculating the signal quality  based on other parameters in the data set 
def calculate_signal_quality(dBm):
    # Normalize dBm: assuming -70 is best, -120 is worst
    quality = (1 - (abs(dBm) - 70) / 50) * 100
    return np.clip(quality, 0, 100)

df["Signal Quality (%)"] = df["Signal Strength (dBm)"].apply(calculate_signal_quality)

# Step 2: Save the updated DataFrame
df.to_csv("signal_metrics_updated.csv", index=False)
```

```python
df.head() # VIEWING THE DATA SET AGAIN TO SEE THE UPDATED SIGNAL QUALITY
```

```python
df.isnull().sum()
```

```python
# ADD A NEW COLUMN AT THE ENDING TELLING IF ITS GOOD TO POOR BASED ON THE VALUES GIVEN 
def categorize_signal(q):
    if q >= 80:
        return "Excellent"
    elif q >= 60:
        return "Good"
    elif q >= 40:
        return "Fair"
    else:
        return "Poor"

df['Signal Category'] = df['Signal Quality (%)'].apply(categorize_signal)
df.head()
```

```python
# Pivot for average Signal Quality per Locality
pivot = df.pivot_table(values='Signal Quality (%)', index='Locality', columns='Network Type', aggfunc='mean')
#index = y axis, columns = x axis, aggerate function type can be sum , mean count etc acc to numpy func
sns.heatmap(pivot, annot=True, cmap='YlGnBu')
plt.title("Signal Quality by Locality & Network Type")
plt.show()
```

```python
# Try to find the time-related column name
# You can change this manually if needed
possible_time_cols = [col for col in df.columns if 'time' in col.lower() or 'date' in col.lower()]
if not possible_time_cols:
    raise ValueError("No column found containing 'time' or 'date'. Please check your CSV columns.")

time_col = possible_time_cols[0]  # Choose the first match

# Convert the time column to datetime
df[time_col] = pd.to_datetime(df[time_col], errors='coerce')

# Drop rows with missing time or signal quality
df = df.dropna(subset=[time_col, 'Signal Quality (%)'])

# we must Sort and set time index
df = df.sort_values(by=time_col)
df.set_index(time_col, inplace=True)

# Plot Signal Quality over time
plt.figure(figsize=(14, 6))
sns.lineplot(data=df, x=df.index, y='Signal Quality (%)')
plt.title("Signal Quality Over Time")
plt.xlabel("Time")
plt.ylabel("Signal Quality (%)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Load and process data
df = pd.read_csv("signal_metrics_updated.csv")
df['Timestamp'] = pd.to_datetime(df['Timestamp'], errors='coerce')
df = df.dropna(subset=['Timestamp', 'Signal Quality (%)'])
#time stamp heatmap
df['Hour'] = df['Timestamp'].dt.hour

# Create pivot: Hour vs. Locality (you can replace 'Locality' with other columns like 'Device ID' or 'Network Type')
pivot = df.pivot_table(values='Signal Quality (%)', index='Locality', columns='Hour', aggfunc='mean')

# Plot heatmap
plt.figure(figsize=(14, 6))
sns.heatmap(pivot, annot=True, fmt=".1f", cmap='YlOrRd')
plt.title("Hourly Average Signal Quality by Locality")
plt.xlabel("Hour of Day")
plt.ylabel("Locality")
plt.tight_layout()
plt.show()
```

```python
# Extract Date and Hour
df['Date'] = df['Timestamp'].dt.date
df['Hour'] = df['Timestamp'].dt.hour

# Pivot by Date (rows) and Hour (columns)
pivot_time = df.pivot_table(values='Signal Quality (%)', index='Date', columns='Hour', aggfunc='mean')

# Plot time series heatmap
plt.figure(figsize=(14, 6))
sns.heatmap(pivot_time, annot=True,fmt=".1f", cmap="coolwarm",cbar_kws={'label': 'Signal Quality (%)'})# if u dont need values remove fmt and annot make it False
plt.title("Signal Quality Heatmap: Hour of Day vs Date")
plt.xlabel("Hour of Day")
plt.ylabel("Date")
plt.tight_layout()
plt.show()
```

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'], errors='coerce')
df = df.dropna(subset=['Timestamp'])

# Extract Date and Hour
df['Date'] = df['Timestamp'].dt.date
df['Hour'] = df['Timestamp'].dt.hour

# Count the number of records (activity) per date and hour
activity_counts = df.groupby(['Date', 'Hour']).size().reset_index(name='Activity Count')
activity_pivot = activity_counts.pivot(index='Date', columns='Hour', values='Activity Count').fillna(0)

# Plot activity heatmap
plt.figure(figsize=(14, 6))
sns.heatmap(activity_pivot, fmt=".1f", cmap='YlOrRd', annot=True, cbar_kws={'label': 'Activity Count'})
plt.title("Activity Heatmap: Number of Records by Hour and Date")
plt.xlabel("Hour of Day")
plt.ylabel("Date")
plt.tight_layout()
plt.show()
```

```python
# Load and process data
df = pd.read_csv("signal_metrics_updated.csv")
df['Timestamp'] = pd.to_datetime(df['Timestamp'], errors='coerce')
df = df.dropna(subset=['Timestamp', 'Signal Quality (%)'])

# Extract hour from timestamp
df['Hour'] = df['Timestamp'].dt.hour

# we are roounding off to the nearest value
df['Throughput Bin'] = df['Data Throughput (Mbps)'].round()  # Or use pd.cut for bins

# Create pivot: Hour vs. Throughput Bin
pivot = df.pivot_table(
    values='Signal Quality (%)',
    index='Hour',
    columns='Throughput Bin',
    aggfunc='mean'
)

plt.figure(figsize=(14, 6))
sns.heatmap(pivot, annot= False,  cmap='YlOrRd')
plt.title("Hourly Average Signal Quality by Data Throughput")
plt.xlabel("Data Throughput (Mbps)")
plt.ylabel("Hour of Day")
plt.tight_layout()
plt.show()
```
